### Ejercicio 2

Si recuerdas, las variables "Admission" y "Major" parecan tener una relacin clara (la distribucin de ratio de admisin variaba claramente entre departamentos o "Majors"). Utiliza el estadstico adecuado (para variables categricas entre s, Chi-2) para comprobar que esa hiptesis de relacin es estadsticamente significativa, tal y como lo hemos hecho en el workout

```python
plot_categorical_relationship_fin(df_admission,"Admission","Major",relative_freq = True, show_values=True )
```

```python
tabla_contingencia = pd.crosstab(df_admission['Admission'], df_admission['Major'])

tabla_contingencia
```

```python
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(tabla_contingencia)


print("Valor Chi-Cuadrado:", chi2)
print("P-Value:", p)
print("Grados de Libertad:", dof)
print("Tabla de Frecuencias Esperadas:\n", expected)
```

Repite el ejercicio anterior para "tip" y "size" considerando esta ltima como "categrica". Haz el analisis bivariante visual previo.

```python
plot_grouped_histograms(df_tips, cat_col="size",num_col="tip", group_size= 3)
```

```python
sizes = df_tips['size'].unique()  # Obtener los valores nicos de la columna categrica, en este caso la compaa rea
tips_por_size = [df_tips[df_tips['size'] == size]['tip'] for size in sizes] # obtenemos los ingresos por compaa y los incluimos en una lista
tips_por_size
```

```python
from scipy import stats
```

```python
f_val, p_val = stats.f_oneway(*tips_por_size) # El mtodo * (igual que cuando vimos *args hace mil aos) 
                                                    # lo que hace es separar todos los elementos de la lista y pasarselos como argumento a la funcin
print("Valor F:", f_val)
print("Valor p:", p_val)
```
