#### #1.2  

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.  

Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)  

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand ITC", tamao 18, color azul acero o "steelblue")
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
df_sorted = df.sort_values(by = "population_total", ascending = False)
most_pop_cities = df_sorted.head(5)
```

```python
most_pop_cities
```

```python
colores = ["blue", "grey", "olive", "brown"]

plt.figure(figsize = (10,5))
bars = plt.bar(most_pop_cities["city"], height = (most_pop_cities["population_total"]/1000).round(3), color = colores, alpha = 0.5)
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 0.2,
        f"{height}",
        ha="center",
        va="bottom"
    )
plt.ylabel("Habitantes (miles)")
plt.ylim(0,5000)
plt.xlabel("Ciudades");
```

```python
plt.hist(df["area_total_km2"], bins = 50, color = "steelblue", alpha = 0.5, label = "Hist Area")
plt.xlabel("Superficie($km^2$)",fontdict= {
    'family': 'Bradley Hand ITC',
    'size': 14,
    'weight': 'bold'} )
plt.ylabel("Numero de ciudades",fontdict= {
    'family': 'Bradley Hand ITC',
    'size': 14,
    'weight': 'bold'} )
plt.legend();
```

```python
plt.scatter(df["area_water_km2"],(df["population_total"]/1000).round(3),color= "green", alpha = 0.5)
plt.xlabel("Area de agua en la ciudad ($km^2$)")
plt.ylabel("Poblacion ( miles habitantes )")

max_agua = df["area_water_km2"].max()
outlier_row = df[df['area_water_km2'] == max_agua]

x1 = outlier_row['area_water_km2'].values[0]
y1 = (outlier_row['population_total'].values[0]/1000).round(3)

max_pop = df["population_total"].max()
outlier_row = df[df['population_total'] == max_pop]

x2 = outlier_row['area_water_km2'].values[0]
y2 = (outlier_row['population_total'].values[0]/1000).round(3)

mytext_xy = (60,3500)
plt.annotate(
    "Outliers",
    xy = (x1,y1),
    xytext = mytext_xy,
    arrowprops= dict(arrowstyle = "->")
)
plt.annotate(
    " ",
    xy = (x2,y2),
    xytext = (mytext_xy[0] + 10,mytext_xy[1] + 50),
    arrowprops= dict(arrowstyle = "->"))
plt.title("Poblacion por kilometro cuadrado de agua");
```

```python
plt.boxplot(df["elevation_m"], vert = False, whis = 1.5, patch_artist = True,
            boxprops=dict(facecolor='red'),
            medianprops = dict(color = "black"))

plt.xlabel("Altitud(m)")
plt.title("Diagrama de caja de la Altitud de las ciudades de California");
```
